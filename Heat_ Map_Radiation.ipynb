```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Raw_Data.csv")
```

```python
DNI = df['DNI'].to_numpy().reshape(365,24).transpose()

fig2, ax2 = plt.subplots(figsize=(12,4))

im2 = ax2.imshow(DNI, cmap='viridis',interpolation='nearest', aspect='auto')
ax2.set_xlabel('Day', fontsize=12)
ax2.set_ylabel('Hour', fontsize=12)
cbar = fig2.colorbar(im2)
#cbar.set_label("DNI [$W/m^2$]", loc='top')

ax3 = cbar.ax
ax3.text(1180,1100,'DNI [$W/m^2$]',rotation=0, fontsize=12)

plt.show()
```
